```python
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time
import json
import re
```

```python
def saveTotxt(disNum, date_times, user1s, user2s, content_goods, title):
    with open('data/' + title + '.txt', 'a+', encoding='utf-8') as f:
        for i in range(date_times.__len__()):
            f.write(str(disNum) + ',')
            f.write(date_times[i] + ',')
            f.write(user1s[i] + ',')
            f.write(user2s[i] + ',')
            f.write(content_goods[i].replace(',','，') + '\n')
            disNum += 1
        f.close()
    return disNum
```

```python
def getFromCurr(disNum, currPage, title, browser):
    print('Working on page ' + str(currPage) + '...')
    while True:
        date_times = []
        user1s = []
        user2s = []
        content_goods = []
        weibos = browser.find_elements_by_css_selector('.list_box .list_ul .list_li')
        i = 0
        while(i < weibos.__len__()):
            try:
                date_time = weibos[i].find_element_by_css_selector('.list_con .WB_func .WB_from').text
                users = weibos[i].find_elements_by_css_selector('.list_con .WB_text a')
                content_bad = weibos[i].find_element_by_css_selector('.list_con .WB_text').text
                user1 = users[0].text
                user2 = ''
                content_good = content_bad.replace(user1+'：', '')
                if users.__len__() > 1:
                    for k in range(users.__len__()):
                        if users[k].text.find('@') != -1:
                            user2 = users[k].text.replace('@', '')
                            content_good = content_good.replace('回复@'+user2+':', '')
                date_times.append(date_time)
                user1s.append(user1)
                user2s.append(user2)
                content_goods.append(content_good)
                i += 1
            except StaleElementReferenceException:
                print('Get a content StaleElementReferenceException')
                break
        if i == weibos.__len__():
            break
    disNum = saveTotxt(disNum, date_times, user1s, user2s, content_goods, title)
    print('Done')
    return disNum
```

```python
def mainWork(disNum, url, title, browser):
    browser.get(url)
    time.sleep(3)
    keyStr = browser.find_element_by_css_selector('.WB_cardpage .W_pages .S_txt1 + a').get_attribute('action-data')
    keyStr_id = re.match('.*?(\d+).*?$', keyStr)
    currPage = 1
    browser.find_element_by_css_selector("a[action-data='id=" + keyStr_id.group(1) + "&page=" + str(currPage) + "']").click()
    disNum = getFromCurr(disNum, currPage, title, browser)
    try:
        maxPage = int(browser.find_element_by_css_selector('.WB_cardpage .W_pages .S_txt2 + a').text)
    except NoSuchElementException:
        maxPage = browser.find_elements_by_css_selector('.WB_cardpage .W_pages a').__len__() - 2
    for i in range(maxPage-1):
        currPage = i + 2
        while True:
            try:
                browser.find_element_by_css_selector("a[action-data='id=" + keyStr_id.group(1) + "&page=" + str(currPage) + "']").click()
                break
            except StaleElementReferenceException:
                print('Get a page StaleElementReferenceException')
            except NoSuchElementException:
                print('Get a page NoSuchElementException')
                time.sleep(0.5)
        time.sleep(1)
        disNum = getFromCurr(disNum, currPage, title, browser)
    print('Finished')
    return disNum
```

```python
def endToStartMW(disNum, url, title, browser):
    browser.get(url)
    time.sleep(3)
    keyStr = browser.find_element_by_css_selector('.WB_cardpage .W_pages .S_txt1 + a').get_attribute('action-data')
    keyStr_id = re.match('.*?(\d+).*?$', keyStr)
    try:
        maxPage = int(browser.find_element_by_css_selector('.WB_cardpage .W_pages .S_txt2 + a').text)
    except NoSuchElementException:
        maxPage = browser.find_elements_by_css_selector('.WB_cardpage .W_pages a').__len__() - 2
    currPage = maxPage
    while(currPage > 0):
        while True:
            try:
                browser.find_element_by_css_selector("a[action-data='id=" + keyStr_id.group(1) + "&page=" + str(currPage) + "']").click()
                break
            except StaleElementReferenceException:
                print('Get a page StaleElementReferenceException')
            except NoSuchElementException:
                print('Get a page NoSuchElementException')
                time.sleep(0.5)
        time.sleep(1)
        disNum = getFromCurr(disNum, currPage, title, browser)
        currPage -= 1
    print('Finished')
    return disNum
```
